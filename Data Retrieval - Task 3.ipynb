```python
# Import libraries for webscraping from  KAYAK Webpage - Edin Kciku (Student 3033225)
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import os
import time


# mapping between destinations and airport codes - Edin Kciku (Student 3033225)  
destinations = {"Edinburgh": "EDI",
                "New York City": "JFK",
                "Miami": "Mia",
                "Barcelona": "BCN",
                "Lisbon": "LIS",
                "Rome": "ROM",
                "Paris": "CDG",
                "Dubai": "DWC",
                "Cape Town": "CPT",
                "Singapore": "SIN",
                "Tokyo": "TYO"}



# create list of destinations - Degnan Kopa (Student 2942646)
cols = ["Edinburgh"]
for key in destinations:
    if key == "Edinburgh":
        continue
    cols.append(key)

# fixed flight date for each destination - remains the same - Degnan Kopa (Student 2942646)
flight_date = "2022-06-01"

# initilize dataframe for flight prizes with empty values - - Degnan Kopa (Student 2942646)
df_flight_prices = pd.DataFrame(columns=cols, index=cols)    
display(df_flight_prices)
```

## Edin Kciku (Student 3033225) 

```python
def get_web_data(city):
    global destinations, cols, df_flight_prices, flight_date
    cwd = os.getcwd()  # get the current working directory  

    # Example: https://www.kayak.co.uk/flights/EDI-JFK/2022-05-16?sort=bestflight_a
    # install chromedriver beforehand and initilize it - Edin Kciku (Student 3033225) 
    path = cwd + "\Downloads\chromedriver_win32\chromedriver.exe"  
    chrome_options = webdriver.ChromeOptions()  # chustomizing options 

    prices = []
    
    # access each row of the empty dataframe by the index / city names 
    for el in cols: # loop over each column / city of the empty dataframe
        row = df_flight_prices.loc[city, :]
        if el == city:
            row[city] = 0 # if the index of the row corresponds with a column name, set prize to 0
        else:
            # Access webpage for each destination
            driver = webdriver.Chrome(path)
            url = "https://www.kayak.co.uk/flights/" + destinations[city] + "-" + destinations[el] + "/" + flight_date + "?sort=bestflight_a"
            driver.get(url) 
            # Set hard time out for the back-end: wait for the page to load
            time.sleep(30)
            # get html content from the webpage
            soup=BeautifulSoup(driver.page_source, "lxml")
            
            # find all available prices on the loaded page
            prices_web = soup.find_all("span", attrs={"class": "price-text"})
            # create a python compatible list of all available prices and
            # clean up the string and transform them into decimal numbers
            for div in prices_web:
                price = div.getText()
                price = price.replace(",", "")
                price = price.replace("£", "")
                price = price.strip()
                price = float(price)
                prices.append(price)

            # only if prizes are found
            if len(prices) > 0:
                min_prize = min(prices)  # retrieve the smalles prize
                # display progress
                print(f"{destinations[city]}-{destinations[el]} {min_prize}")
                row[el] = min_prize # fill up accessed cell of the empty dataframe

            prices.clear()
            driver.close()  # close the web session


    
```

## Degnan Kopa (Student 2942646)

```python
# for each starting city - retrieve the minimal prizes to all cities
cities = cols.copy()
for city in cities:
    get_web_data(city)
```
