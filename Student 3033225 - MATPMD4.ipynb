```python
import random 
import numpy as np
import math

# initilize random starting points for x, y, and z 
# with uniform distribution between -1 and 1
x = random.uniform(-1, 1)
y = random.uniform(-1, 1)
z = random.uniform(-1, 1)

# find the best values of initial x, y, and z
fbest = math.exp(math.sin(40*z)) + math.exp(math.sin(50*x)) + math.sin(60*math.exp(y))
+ math.sin(70*math.sin(x)) + math.sin(70*math.cos(z)) + math.sin(math.sin(80*y))
- math.sin(10*(x+y)) + (x**2+y**2+z**2)/4

steps = 10000
# calculate steps times new random values for x, y, and z
# with uniform distribution and compare the derived value funtion with
# it predecessor
for i in range(1,steps):
    xn = random.uniform(-1, 1)
    yn = random.uniform(-1, 1)
    zn = random.uniform(-1, 1)
    fnew = math.exp(math.sin(40*zn)) + math.exp(math.sin(50*xn)) + math.sin(60*math.exp(yn))+ math.sin(70*math.sin(xn)) + math.sin(70*math.cos(zn)) + math.sin(math.sin(80*yn)) - math.sin(10*(xn+yn)) + (xn**2+yn**2+zn**2)/4
    
    # if the new function value is greate than its predecessor
    # change the values of the best function value with the new one
    # and change the old values of x, y and z with the new ones
    if fnew > fbest:
        x = xn
        y = yn
        z = zn
        fbest = fnew

print ('xbest' , x, 'ybest', y, 'zbest', z, 'fbest', fbest,)
```

```python
best_prices = []
best_route = []
#columns = list(df_flights.columns)
start = "Edinburgh"
visited = []
df_copy = df_flights.copy()
columns = list(df_flights.columns)

# loop over dataframe "number of rows" times - but: not in the order of the indexes
# since for each iteration, we retrieve the column name / city of the cheapest price
# and then go to the corresponding index in the next iteration
for i in range(0, len(df_flights.index)):  
    best_price = 0
    best_dest = None
    row = df_copy.loc[index_dest[start],:]
    for col in columns:
        if start == col or col in visited: # already visited city cannot be visited again
            continue
        else:
            if best_price == 0 or best_price > row[col]: 
                
                best_price = row[col]
                best_dest = col
    # if number of destination is 10 (maximum) travel back to Edinburgh
    if len(best_route) == 10:  
        best_prices.append(row["Edinburgh"])
        best_route.append("Edinburgh")
        continue
    visited.append(start)
    best_prices.append(best_price)
    best_route.append(best_dest)
    start = best_dest
    best_dest = None
    best_price = 0
            
            
print(best_prices)
print(best_route)
```
